```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
std_perf = pd.read_csv("StudentsPerformance.csv")

std_perf_gender = std_perf.groupby(['gender']).mean()
std_perf_gender["Total"] = 0.00
std_perf_gender["Total"] = (std_perf_gender["math score"] + std_perf_gender["reading score"] + std_perf_gender["writing score"])/3
std_perf_gender = std_perf_gender.reset_index()
```

```python
plt.figure(figsize=(15,15))
plt.subplot(4,1,1)
plt.tight_layout()
plt.bar(std_perf_gender["gender"], std_perf_gender["Total"], color='darkblue')
plt.xlabel("Gender")
plt.ylabel("Mean of the total earned")
plt.xticks(range(2), std_perf_gender["gender"], fontsize=14)
plt.yticks(std_perf_gender["Total"], fontsize=14)
plt.title("Mean total marks earned by each gender")
for i in range(0,2):
    plt.text(std_perf_gender["gender"][i], std_perf_gender["Total"][i]+0.5, str(std_perf_gender["Total"][i]), fontsize=10)
plt.subplot(4,1,2)
plt.tight_layout()
plt.bar(std_perf_gender["gender"], std_perf_gender["math score"], color='blue')
plt.xlabel("Gender")
plt.ylabel("Mean of the maths score earned")
plt.xticks(range(2), std_perf_gender["gender"], fontsize=14)
plt.yticks(std_perf_gender["math score"], fontsize=14)
plt.title("Mean maths scores marks earned by each gender")
for i in range(0,2):
    plt.text(std_perf_gender["gender"][i], std_perf_gender["math score"][i]+0.5, str(std_perf_gender["math score"][i]), fontsize=10)
plt.subplot(4,1,3)
plt.tight_layout()
plt.bar(std_perf_gender["gender"], std_perf_gender["reading score"], color='red')
plt.xlabel("Gender")
plt.ylabel("Mean of the reading score earned")
plt.xticks(range(2), std_perf_gender["gender"], fontsize=14)
plt.yticks(std_perf_gender["reading score"], fontsize=14)
plt.title("Mean reading score marks earned by each gender")
for i in range(0,2):
    plt.text(std_perf_gender["gender"][i], std_perf_gender["reading score"][i]+0.5, str(std_perf_gender["reading score"][i]), fontsize=10)
plt.subplot(4,1,4)
plt.tight_layout()
plt.bar(std_perf_gender["gender"], std_perf_gender["writing score"], color='brown')
plt.xlabel("Gender")
plt.ylabel("Mean of the writing score earned")
plt.xticks(range(2), std_perf_gender["gender"], fontsize=14)
plt.yticks(std_perf_gender["writing score"], fontsize=14)
plt.title("Mean writing score marks earned by each gender")
for i in range(0,2):
    plt.text(std_perf_gender["gender"][i], std_perf_gender["writing score"][i]+0.5, str(std_perf_gender["writing score"][i]), fontsize=10)
plt.show()
```
